```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Φόρτωση του CSV
df = pd.read_csv('sensor_data.csv')

# Μετατροπή της ώρας σε σωστή μορφή
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df = df.sort_values('Timestamp').reset_index(drop=True)

# Γέμισμα τυχόν κενών τιμών
df['Humidity'] = df['Humidity'].ffill()

print(f"Συνολικές μετρήσεις: {len(df)}")
df.head()
```

```python
print("--- Στατιστική Ανάλυση ---")
display(df.describe())

plt.figure(figsize=(12, 5))
plt.plot(df['Timestamp'], df['Humidity'], color='blue')
plt.title('Ιστορικά Δεδομένα Υγρασίας (Raw Values)')
plt.xlabel('Χρόνος')
plt.ylabel('Τιμές')
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

print("--- Μοντέλο: Linear Regression ---")

# Count ως X και την Υγρασία ως y
X = df[['Count']].values
y = df['Humidity'].values

# Εκπαίδευση
lin_model = LinearRegression()
lin_model.fit(X, y)

# Πρόβλεψη για τα επόμενα 5 πακέτα
last_count = df['Count'].max()
future_counts = np.array([[last_count + 1], [last_count + 2], [last_count + 3], [last_count + 4], [last_count + 5]])

predictions = lin_model.predict(future_counts)

for i, pred in enumerate(predictions):
    print(f"Μελλοντικό Πακέτο {int(future_counts[i][0])}: Προβλεπόμενη Υγρασία = {pred:.2f}")
```

```python
from statsmodels.tsa.arima.model import ARIMA
import warnings
warnings.filterwarnings("ignore")

print("--- Μοντέλο: ARIMA ---")

# Για να μην καθυστερήσει πολύ το ARIMA,το εκπαιδεύουμε στις τελευταίες 500 μετρήσεις.
time_series = df['Humidity'].values[-500:]

# Εκπαίδευση με παραμέτρους (1, 1, 1)
arima_model = ARIMA(time_series, order=(1, 1, 1))
arima_result = arima_model.fit()

# Πρόβλεψη 5 επόμενων βημάτων
forecast = arima_result.forecast(steps=5)

for i, pred in enumerate(forecast):
    print(f"Μελλοντικό Βήμα {i+1}: Προβλεπόμενη Υγρασία = {pred:.2f}")
```
